```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Optional: display plots in notebook
%matplotlib inline

# Set seaborn style
sns.set(style="whitegrid")
df = pd.read_csv(r"C:\Users\vedan\OneDrive\Desktop\train.csv")
df.head()

```

```python
## Step 2 – Basic Data Exploration
df.info()
df.describe()
df.isnull().sum()
df['Survived'].value_counts()
```

```python
df['Age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title('Fare Distribution')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.show()
```

```python

numeric_df = df.select_dtypes(include=['float64', 'int64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.show()
```

```python
# Summary of Findings

# - Gender: Females had a much higher survival rate than males.
# - Passenger Class: 1st class passengers survived more often compared to 2nd and 3rd class.
# - Fare: Higher fares were positively correlated with survival, indicating wealthier passengers had better chances.
# - Age: Younger passengers had slightly better survival rates, but age impact was weaker compared to class and gender.
# - Correlation: Survival showed a strong positive correlation with being in higher class and female gender, and a negative correlation with Pclass (lower classes).

```
